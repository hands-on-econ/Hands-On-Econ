```r
data<-read.csv('https://github.com/hands-on-econ/Hands-On-Econ/raw/master/Datasets/countries%20of%20the%20world.csv')
data[,'lGDPcap']<-log(data[,'GDP....per.capita.'])
data[,'Literacy....']<-as.numeric(gsub(',','',data[,'Literacy....']))/10
data[,'Literacy....']<-as.numeric(gsub(',','',data[,'Agriculture']))/10
data[,'Literacy....']<-as.numeric(gsub(',','',data[,'Service']))/10
data2<-na.omit(data[c('lGDPcap','Literacy....')])


```

```r
library(ggplot2)
data2[,'Residuals']<-lm(data2[,'lGDPcap']~data2[,'Literacy....'])$residuals
ggplot(data2,aes(Literacy....,Residuals))+geom_point(colour='lawngreen')+ggtitle('Residual Plot')
```

```r
library(gridExtra)
options(repr.plot.width=8, repr.plot.height=5)
data3<-na.omit(data[c('lGDPcap','Literacy....','Agriculture')])
data3[,'Residuals']<-lm(data3[,'lGDPcap']~data3[,'Literacy....']+data3[,'Agriculture'])$residuals
y1<-ggplot(data3,aes(Literacy....,Residuals))+geom_point(colour='indianred')+ggtitle('Residual Plot (Multivariate)')
y2<-ggplot(data2,aes(Literacy....,Residuals))+geom_point(colour='lawngreen')+ggtitle('Residual Plot (Bivariate)')
grid.arrange(y1,y2,ncol=1)


```

```r
data3<-na.omit(data[c('lGDPcap','Literacy....','Agriculture')])
data3[,'Residuals']<-lm(data3[,'lGDPcap']~data3[,'Literacy....']+data3[,'Agriculture'])$residuals
data3[,'Residuals1']<-lm(data3[,'lGDPcap']~data3[,'Literacy....'])$residuals
y1<-ggplot(data3,aes(Literacy....,Residuals))+geom_point(colour='indianred')+ggtitle('Residual Plot (Multivariate)')
y2<-ggplot(data3,aes(Literacy....,Residuals1))+geom_point(colour='lawngreen')+ggtitle('Residual Plot (Bivariate)')
grid.arrange(y1,y2,ncol=1)



```
