```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
data=pd.read_csv('https://github.com/hands-on-econ/Hands-On-Econ/raw/master/Datasets/acs2015_county_data.csv')
data.head(20)
```

```python
fig=plt.figure(figsize=(12,8),dpi=300)
ax = fig.add_subplot(2, 2, 1)
sns.distplot(data['Poverty'],color='blue')
ax = fig.add_subplot(2, 2, 2)
sns.distplot(data['Professional'],color='green')
ax = fig.add_subplot(2, 2, 3)
sns.distplot(data['Unemployment'],color='red')
ax = fig.add_subplot(2, 2, 4)
sns.distplot(data['IncomePerCap'],color='orange')
plt.show()

```

```python
plt.figure(figsize=(10, 7),dpi=300)
ax=sns.scatterplot(data['Men'],data['Women'])
coef=data[['Men','Women']].corr().iloc[1,0]
ax.text(3500000,1500000,'$\\rho$='+str(round(coef,4)))
plt.title('Simple Gender Scatterplot')
plt.xlabel('Number of Men')
plt.ylabel('Number of Women')
plt.show()
```

```python
corrmat=data.corr()
plt.figure(figsize=(10,7),dpi=300)
sns.heatmap(corrmat)
plt.title('Correlation Heatmap for US Census Data')
plt.show()
```
